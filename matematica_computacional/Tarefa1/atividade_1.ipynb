2- Inicialmente calculamos o valor do determinante antes das transformações:
$$
\begin{bmatrix}
1&0\\
0&1\\
\end{bmatrix}
$$
det(A) = (1.1) - (0.0) = 1

é importante ressaltar que o determinante representa a área da figura, a qual pode ser observada visualmente.

```python
import matplotlib.pyplot as plt
import numpy as np
#criando o vetor i
origem_i = [0,0]
final_i = [1,0]
plt.quiver(*origem_i,*final_i, angles = "xy", scale_units = "xy", scale = 1, color = "green") 
#criando o vetor j
origem_j = [0,0]
final_j = [0,1]
plt.quiver(*origem_j,*final_j, angles = "xy", scale_units = "xy", scale = 1, color =  "red")
#Criando as legendas para cada vetor
plt.figtext(0.5,0.65, "Vetor J")
plt.figtext(0.65,0.5, "Vetor i")
#Mostrando a área
vi = np.array([1,0])
vj = np.array([0,1])
def determinante(v1, v2):
    matrix = np.array([v1, v2]).T
    return np.linalg.det(matrix)
corners = np.array([[0,0], vi, vi+vj, vj, [0,0]])
plt.fill(corners[:,0], corners[:,1], color='green', alpha=0.3,label=f"det = {abs(determinante(vi,vj)):.2f}")
#Organizando o gráfico
plt.grid(True)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

5- transformação linear em um vetor x de duas dimensões para obter um vetor y orientado a 90° em relação a x. Escolhendo um vetor X:


```python
matriz1 = np.array([[5],[7]])
matrizRotacao = np.array([[0,1],[-1,0]])
matriz_final = matrizRotacao  @ matriz1
print(matriz_final)
```

Utilizando o vetor
$$
X
=
\begin{bmatrix}
5\\
7\\
\end{bmatrix}
$$

e multiplicando pela matriz de rotação:
$$
\begin{bmatrix}
0 & 1 \\
-1 & 0\\
\end{bmatrix}
$$

obtemos o vetor Y:
$$
Y
=
\begin{bmatrix}
7 \\
-5 \\
\end{bmatrix}
$$

calculando o produto escalar, obtemos:

X ⋅ Y = (5.7) + (7.-5) = 35 + (-35) = 0

Como o produto escalar entre os vetores resulta em 0, o que indica que a transformação linear gera um vetor a 90° do vetor original e para quaisquer valores do vetor X o vetor Y será perpendicular

```python
import matplotlib.pyplot as plt
import numpy as np
#criando o vetor i
plt.quiver(5,7, angles = "xy", scale_units = "xy", scale = 1, color = "green") 
#criando o vetor j
plt.quiver(7,-5, angles = "xy", scale_units = "xy", scale = 1, color =  "red")
#Criando as legendas para cada vetor
plt.figtext(0.65,0.8, "Vetor X")
plt.figtext(0.75,0.35, "Vetor Y")
#Organizando o gráfico
plt.grid(True)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

6-Analisando resultados do produto dot e conclusão

```python
import numpy as np

v1 = np.array([1, 2])
v2 = np.array([4, 5])

dot = np.dot(v1, v2)

print(dot)
```

O produto dot entre os vetores resultou num inteiro possitivo, o que indica que os vetores não são perpendiculares já que o produto é diferente de 0 e indica que os vetores apontam para direções similares.
