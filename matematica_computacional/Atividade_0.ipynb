1 - Rotação horária de 90° e shear (cisalhamento) em x, diferentes dos exemplos apresentados em sala.

Inicialmente, irei criar o gráfico dos vetores unitários i e j utilizando a biblioteca matplotlib. Os valores de cada vetor foram escolhidos de uma forma que facilite sua visualização

```python
import matplotlib.pyplot as plt
import numpy as np
#criando o vetor i
origem_i = [0,0]
final_i = [1,0]
plt.quiver(*origem_i,*final_i, angles = "xy", scale_units = "xy", scale = 1, color = "green") 
#criando o vetor j
origem_j = [0,0]
final_j = [0,1]
plt.quiver(*origem_j,*final_j, angles = "xy", scale_units = "xy", scale = 1, color =  "red")
#Criando as legendas para cada vetor
plt.figtext(0.5,0.65, "Vetor J")
plt.figtext(0.65,0.5, "Vetor i")
#Organizando o gráfico
plt.grid(True)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

Em seguida, alteramos os valores dos vetores para obter a sua rotação 

```python
#criando o vetor i
plt.quiver(0,0,0,-1, angles = "xy", scale_units = "xy", scale = 1, color = "green")
#criando o vetor j
plt.quiver(0,0,1,0, angles = "xy", scale_units = "xy", scale = 1, color =  "red")
#Criando as legendas para cada vetor
plt.figtext(0.65,0.5, "Vetor J")
plt.figtext(0.5,0.3, "Vetor i")
#Organizando o gráfico
plt.grid(True)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

Agora, para mostrar o cisalhamento dos vetores, vamos multiplicar o vetor j (0,2) pela constante k = 2, obtendo.

```python
#criando o vetor i
k = 2
origem_i = [0,0]
final_ix = 1
final_iy = 0
final_iCisalhamento = [0 + k*final_ix, final_iy]
plt.quiver(*origem_i,*final_iCisalhamento, angles = "xy", scale_units = "xy", scale = 1, color = "green") 
#criando o vetor j
origem_j = [0,0]
final_jX = 0
final_jY = 1
final_jCisalhamento = [0 + k*final_jY, final_jY]
plt.quiver(*origem_j,*final_jCisalhamento, angles = "xy", scale_units = "xy", scale = 1, color =  "red")
#Criando as legendas para cada vetor
plt.figtext(0.7,0.6, "Vetor J")
plt.figtext(0.65,0.4, "Vetor i")
#Organizando o gráfico
plt.grid(True)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

2- A obtenção das matrizes de transformação a partir dos vetores unitários (bases) — mostrando o processo numericamente e graficamente com o Matplotlib e aplicação passo a passo dessas matrizes a dois vetores arbitrários, com apresentação de resultados numéricos e gráficos no Matplotlib.

Usando os vetores unitários  
$$
\vec{i} =
\begin{bmatrix}
1 \\
0 \\
\end{bmatrix}
$$
$$
\vec{j} =
\begin{bmatrix}
0 \\
1 \\
\end{bmatrix}
$$
obtemos a matriz 2x2 com os vetores unitários
$$
\begin{bmatrix}
1&0\\
0&1\\
\end{bmatrix}
$$
Para obter a matriz de transformação da rotação em 90° mostrada anteriormente vamos concatenar os vetores já rotacionados
$$
\vec{i} =
\begin{bmatrix}
0 \\
-1 \\
\end{bmatrix}
$$
$$
\vec{j} =
\begin{bmatrix}
1 \\
0 \\
\end{bmatrix}
$$
obtendo assim a matriz
$$
\begin{bmatrix}
0&1\\
-1&0\\
\end{bmatrix}
$$
É possivel aplicar essa rotação a qualquer par de vetores, por exemplo os vetores: 

$$
\vec{A} =
\begin{bmatrix}
3 \\
3 \\
\end{bmatrix}
$$
$$
\vec{B} =
\begin{bmatrix}
5 \\
9 \\
\end{bmatrix}
$$

multiplicamos os valores pela matriz de rotação:
$$
\begin{bmatrix}
3 \\
3
\end{bmatrix}
\cdot
\begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
=
\begin{bmatrix}
3 \\
-3
\end{bmatrix}
$$
$$
\begin{bmatrix}
5 \\
9
\end{bmatrix}
\cdot
\begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
=
\begin{bmatrix}
9 \\
-5
\end{bmatrix}
$$
$$
\vec{A} =
\begin{bmatrix}
3&9 \\
-3&-5 \\
\end{bmatrix}
$$

```python
import matplotlib.pyplot as plt
#criando o vetor A
origem_i = [0,0]
final_i = [3,-3]
plt.quiver(*origem_i,*final_i, angles = "xy", scale_units = "xy", scale = 1, color = "green") 
#criando o vetor B
origem_j = [0,0]
final_j = [9,-5]
plt.quiver(*origem_j,*final_j, angles = "xy", scale_units = "xy", scale = 1, color =  "red")
#Criando as legendas para cada vetor
plt.figtext(0.7,0.25, "Vetor B rotacionado")
plt.figtext(0.55,0.35, "Vetor A rotacionado")
#Organizando o gráfico
plt.grid(True)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

Agora, faremos o mesmo processo com o cisalhamento, aplicando ele a dois vetores quaisquer:
$$
\vec{A} =

3 \\
\end{bmatrix}
$$
$$
\vec{B} =

T = (x + k*Y,Y)

TA = (3 + 2*3, 3) = (9,3)

TB = (5 + 2*9, 9)= (23,9)

$$
\vec{Matrizcisalhamento} =
\begin{bmatrix}

```python
#criando o vetor A
origem_i = [0,0]
final_i = [9,3]
plt.quiver(*origem_i,*final_i, angles = "xy", scale_units = "xy", scale = 1, color = "green") 
#criando o vetor B
origem_j = [0,0]
final_j = [23,9]
plt.quiver(*origem_j,*final_j, angles = "xy", scale_units = "xy", scale = 1, color =  "red")
#Criando as legendas para cada vetor
plt.figtext(0.8,0.7, "Vetor B")
plt.figtext(0.6,0.5, "Vetor A")
#Organizando o gráfico
plt.grid(True)
plt.xlim(-25,25)
plt.ylim(-25,25)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```
